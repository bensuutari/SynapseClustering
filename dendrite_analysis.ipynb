```python
import pandas as pd
import os
```

```python
data=pd.read_csv(os.getcwd()+'/data/xrzct.csv')
data.drop(['Comment'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data['DendriteID'] = data['Trace'].astype(str).str[0:4]
```

```python
data['SynapseID'] = data['Trace'].astype(str).str[5:]
```

```python
data.to_csv(os.getcwd()+'/data/xrzctBenParse.csv')
```

```python
datasumarea=data.groupby(['DendriteID']).sum()['Area']
```
